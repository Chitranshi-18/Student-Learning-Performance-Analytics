```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('../data/merged_dataset_cleaned.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numeric columns for plotting
num_cols = ['StudyHours','Attendance','Extracurricular','AssignmentCompletion',
            'OnlineCourses','Discussions','Motivation','StressLevel','ExamScore','FinalGrade']
```

```python
# Histograms
df[num_cols].hist(figsize=(15,10), bins=20)
plt.suptitle("Distributions of Student Metrics")
plt.show()
```

```python
# Boxplots to check outliers
plt.figure(figsize=(15,6))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot of Student Metrics")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Engagement and Performance")
plt.show()
```

```python
# Engagement vs Performance
plt.figure(figsize=(10,6))
sns.scatterplot(x='StudyHours', y='FinalGrade', hue='LearningStyle', data=df)
plt.title("Study Hours vs FinalGrade by LearningStyle")
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='AssignmentCompletion', y='FinalGrade', hue='LearningStyle', data=df)
plt.title("Assignment Completion vs FinalGrade by LearningStyle")
plt.show()
```

```python
df['EngagementScore'] = df[['StudyHours','Attendance','AssignmentCompletion','OnlineCourses','Discussions']].sum(axis=1)

plt.figure(figsize=(10,6))
sns.histplot(df['EngagementScore'], bins=30, kde=True)
plt.title("Distribution of Composite Engagement Score")
plt.show()

# Engagement vs Performance
plt.figure(figsize=(10,6))
sns.scatterplot(x='EngagementScore', y='FinalGrade', data=df)
plt.title("Engagement Score vs FinalGrade")
plt.show()
```

```python
df.to_csv('../data/merged_dataset_final.csv', index=False)
```
